```python
import matplotlib.pyplot as plt

x = np.arange(0,10,0.01)
y = np.sin(x)
y2 = np.cos(x)
y3 = 0.01*x*x
plt.plot(x,y)
plt.plot(x,y2)
plt.plot(x,y3)

plt.show()
```
